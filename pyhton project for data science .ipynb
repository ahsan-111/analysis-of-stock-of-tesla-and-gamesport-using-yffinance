```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
tesla = yf.Ticker("TSLA")
```

```python
tesla_share_price_history = tesla.history(period="max")
```

```python

tesla_share_price_history.reset_index(inplace=True)
tesla_data = tesla_share_price_history.head()
tesla_data
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkPY0220ENSkillsNetwork23455606-2022-01-01"
```

```python
data = requests.get(url).text
soup = BeautifulSoup(data,"html.parser")
tables = soup.find_all("table")
```

```python
tesla_revenue = pd.DataFrame(columns = ["Date","Revenue"])
for row in tables[1].find_all("tr"):
    col = row.find_all("td")
    if col != []:
        date = col[0].text
        revenue = col[1].text
        tesla_revenue = tesla_revenue.append({"Date":date,"Revenue":revenue},ignore_index=True)
tesla_revenue
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"")
b =tesla_revenue["Revenue"]
b
```

```python
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue
```

```python
tesla_revenue.tail()
```
