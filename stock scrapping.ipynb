```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.x-rates.com/table/?from=PKR&amount=1"
```

```python
data = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html.parser")
soup
```

```python
tables = soup.find_all("table")
```

```python
len(tables)
```

```python
tables[0].text
```

```python
data1 = pd.DataFrame(columns=["Pakistani Rupee","1.00 PKR","inv. 1.00 PKR"])
for row in tables[0].find_all("tr"):
    col = row.find_all("td")
    if col!=[]:
        p = col[0].text
        pk = col[1].text
        inv = col[2].text
        data1 = data1.append({"Pakistani Rupee":p,"1.00 PKR":pk,"inv. 1.00 PKR":inv},ignore_index=True)
data1
        
```
